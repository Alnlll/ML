* $\mathit{ X = \{x^{(1)},x^{(2)},...,x^{(m)}\} }$
![](./data/image/data.png '输入数据')

K-means算法：
* 随机初始化聚类中心 $\mu^{(i)},...,\mu^{(K)}$
* repeat {
    * 根据数据点到聚类中的距离进行聚类分配,$\mathit{ c^{(i)}}$
    * 根据聚类中心被分配的数据点均值更新聚类中心
}
